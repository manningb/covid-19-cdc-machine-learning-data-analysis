```python
def printMetrics(testActualVal, predictions):
    #classification evaluation measures
    print('\n==============================================================================')
    print("MAE: ", metrics.mean_absolute_error(testActualVal, predictions))
    #print("accuracy score:", metrics.accuracy_score(testActualVal, predictions))
    #print("MSE: ", metrics.mean_squared_error(testActualVal, predictions))
    print("RMSE: ", metrics.mean_squared_error(testActualVal, predictions)**0.5)
    print("R2: ", metrics.r2_score(testActualVal, predictions))
    return {"MAE":metrics.mean_absolute_error(testActualVal, predictions), "RMSE":metrics.mean_squared_error(testActualVal, predictions)**0.5, "R2":metrics.r2_score(testActualVal, predictions)}
```

```python
def get_train_test(X, y, test_size, features=None):
    if features == None:
        features = X.columns
    X_train, X_test, y_train, y_test = train_test_split(X[[features]], y, test_size=0.1, random_state=0)
#     print(X[features])
    X_train_dummies = pd.get_dummies(X_train ,drop_first=True)
    y_train_dummies = pd.get_dummies(y_train,drop_first=True)
    X_test_dummies = pd.get_dummies(X_test, drop_first=True)
    X_test_dummies = X_test_dummies.assign(**{col:0 for col in X_train_dummies.columns.difference(X_test_dummies.columns).tolist()})
    y_test_dummies = pd.get_dummies(y_test, drop_first=True)
    return X_train_dummies, X_test_dummies, y_train_dummies, y_test_dummies
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from numpy import set_printoptions

test = SelectKBest(score_func=f_classif, k=4)
fit = test.fit(X_train_dummies, y_train_dummies)
# summarize scores
set_printoptions(precision=3)
print(fit.scores_)
features = list(zip(X_train_dummies.columns, fit.scores_))
coefficients = pd.concat([pd.DataFrame(X_train_dummies.columns, columns=['feature']),pd.DataFrame(np.transpose(fit.scores_), columns=['death_Yes'])], axis = 1,)

print(coefficients.sort_values(by=["death_Yes"], ascending=False).head(15))
features = fit.transform(X_train_dummies)
# summarize selected features
print(features[0:5,:])
```
